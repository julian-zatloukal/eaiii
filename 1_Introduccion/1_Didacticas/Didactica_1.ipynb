La f√≥rmula de Friis se utiliza para calcular el factor de ruido total de etapas en serie, cada una con sus respectivas p√©rdidas o ganancias y su respectivo factor de ruido. donde $F_{resto}$ es el factor de ruido total de las etapas subsecuentes. De acuerdo a la ecuaci√≥n, la cifra de ruido total, $F_{receptor}$, es dominada por la cifra de ruido del amplificador de bajo ruido, $F_{LNA}$, si la ganancia es lo suficientemente alta.
